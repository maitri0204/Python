```python
import pandas as pd
pop = pd.read_csv('population.csv')
print(pop.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pop = pop[['Country Name', '2020']]

top_10_countries = pop.sort_values(by='2020', ascending=False).head(10)
print(top_10_countries)

plt.bar(top_10_countries['Country Name'], top_10_countries['2020'])
plt.xticks(rotation=90)
plt.show()
```
