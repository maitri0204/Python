```python
import pandas as pd
import os
import json
```

```python
with open('Covid19_Cases/Antarctica/Antarctica.json') as file:
    data = json.load(file)
    Antarctica = pd.DataFrame(data)
print(Antarctica)

with open('Covid19_Cases/Asia/China.json') as file:
    data = json.load(file)
    China = pd.DataFrame(data)
print(China)

with open('Covid19_Cases/Asia/India.json') as file:
    data = json.load(file)
    India = pd.DataFrame(data)
print(India)

with open('Covid19_Cases/Europe/England.json') as file:
    data = json.load(file)
    England = pd.DataFrame(data)
print(England)

with open('Covid19_Cases/Europe/France.json') as file:
    data = json.load(file)
    France = pd.DataFrame(data)
print(France)

with open('Covid19_Cases/North_America/Canada.json') as file:
    data = json.load(file)
    Canada = pd.DataFrame(data)
print(Canada)

with open('Covid19_Cases/North_America/USA.json') as file:
    data = json.load(file)
    USA = pd.DataFrame(data)
print(USA)
```

```python
combined_df = pd.concat([Antarctica, China, India, England, France, Canada, USA], ignore_index=True)
combined_df
```

```python
combined_df.isna().sum()
```

```python
analysis = combined_df.groupby('country').agg(total_confirmed=('confirmed_cases', 'sum'), total_deaths=('deaths', 'sum'), total_recovered=('recovered', 'sum'))
analysis['active_cases'] = analysis['total_confirmed'] - analysis['total_deaths'] - analysis['total_recovered']
print(analysis)
```

```python
highest_cases = analysis.sort_values('total_confirmed', ascending=False).head(5)
highest_cases
```

```python
lowest_cases = analysis.sort_values('total_confirmed', ascending=True).head(5)
lowest_cases
```

```python
summary = analysis.to_dict()
with open('covid19_summary.json', 'w') as file:
    json.dump(summary, file)
```
